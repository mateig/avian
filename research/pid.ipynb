```python
import ast
from collections import defaultdict

import matplotlib.pyplot as plt
import numpy as np
```

```python
filepath = "data/flight/rc.txt"
data = defaultdict(list)

with open(filepath, "r") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        try:
            row = ast.literal_eval(line)
            for key, value in row.items():
                data[key].append(value)
        except:
            continue

throttle = np.array(data["throttle"])
roll = np.array(data["roll"])
pitch = np.array(data["pitch"])
yaw = np.array(data["yaw"])

print(f"Loaded {len(throttle)} samples")
```

```python
plt.figure(figsize=(10, 5))
plt.hist(throttle, bins=50, edgecolor='black', alpha=0.7)
plt.axvline(throttle.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {throttle.mean():.0f}')
plt.axvline(np.median(throttle), color='orange', linestyle='--', linewidth=2, label=f'Median: {np.median(throttle):.0f}')
plt.xlabel('Throttle Value')
plt.ylabel('Frequency')
plt.title('Throttle Distribution (All Values)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.plot(throttle, linewidth=0.5, alpha=0.8)
plt.axhline(throttle.mean(), color='red', linestyle='--', linewidth=1, label=f'Mean: {throttle.mean():.0f}')
plt.xlabel('Sample Index')
plt.ylabel('Throttle Value')
plt.title('Throttle Over Time')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

axes[0].hist(roll, bins=30, edgecolor='black', alpha=0.7, color='blue')
axes[0].axvline(roll.mean(), color='red', linestyle='--', label=f'Mean: {roll.mean():.0f}')
axes[0].set_title('Roll')
axes[0].legend()

axes[1].hist(pitch, bins=30, edgecolor='black', alpha=0.7, color='red')
axes[1].axvline(pitch.mean(), color='blue', linestyle='--', label=f'Mean: {pitch.mean():.0f}')
axes[1].set_title('Pitch')
axes[1].legend()

axes[2].hist(yaw, bins=30, edgecolor='black', alpha=0.7, color='green')
axes[2].axvline(yaw.mean(), color='red', linestyle='--', label=f'Mean: {yaw.mean():.0f}')
axes[2].set_title('Yaw')
axes[2].legend()

plt.tight_layout()
plt.show()
```

```python
flying_throttle = throttle[throttle > 1250]

plt.figure(figsize=(10, 5))
counts, bins, _ = plt.hist(flying_throttle, bins=30, edgecolor='black', alpha=0.7, color='green')
plt.axvline(flying_throttle.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {flying_throttle.mean():.0f}')
plt.axvline(np.median(flying_throttle), color='orange', linestyle='--', linewidth=2, label=f'Median: {np.median(flying_throttle):.0f}')

# Mode - most common bin, likely hover throttle
max_bin_idx = np.argmax(counts)
hover_estimate = (bins[max_bin_idx] + bins[max_bin_idx + 1]) / 2
plt.axvline(hover_estimate, color='blue', linestyle='-', linewidth=2, label=f'Mode (Hover): {hover_estimate:.0f}')

plt.xlabel('Throttle Value')
plt.ylabel('Frequency')
plt.title('Throttle Distribution (Flying Only, >1000)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Estimated hover throttle: {hover_estimate:.0f}")
```

```python
print("=" * 60)
print("STATISTICS")
print("=" * 60)
print(f"\nThrottle (flying only):")
print(f"  Min: {flying_throttle.min()}, Max: {flying_throttle.max()}")
print(f"  Mean: {flying_throttle.mean():.1f}, Std: {flying_throttle.std():.1f}")
print(f"  Range used: {flying_throttle.max() - flying_throttle.min()}")

print(f"\nRoll:  Min={roll.min()}, Max={roll.max()}, Mean={roll.mean():.0f}, Std={roll.std():.1f}")
print(f"Pitch: Min={pitch.min()}, Max={pitch.max()}, Mean={pitch.mean():.0f}, Std={pitch.std():.1f}")
print(f"Yaw:   Min={yaw.min()}, Max={yaw.max()}, Mean={yaw.mean():.0f}, Std={yaw.std():.1f}")

print("\n" + "=" * 60)
print("SUGGESTED VALUES")
print("=" * 60)
print(f'\nRC_CENTER = {{"roll": {int(roll.mean())}, "pitch": {int(pitch.mean())}, "throttle": {int(hover_estimate)}, "yaw": {int(yaw.mean())}}}')

suggested_range = int(max(flying_throttle.std() * 2, 20))
print(f'RC_RANGE = {suggested_range}')
```
